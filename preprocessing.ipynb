```python
import pandas as pd
import numpy as np

data = pd.read_csv('dataRawHeaders.csv')

# rename columns with underscores replaced by spaces
data = data.rename(columns=lambda x: x.replace('_', ' '))

# convert date column to datetime format
data['Date'] = pd.to_datetime(data['Date'].astype(str), format='%Y%m%d')
data
```

```python
scores = data.copy()
scores = scores[['Date','Visiting Team', 'Home Team','Visiting Team Score','Home Team Score']]
scores['Home Diff'] = scores['Home Team Score'] - scores['Visiting Team Score']
scores['Home Win'] = scores['Home Diff'] > 0
scores['Home Win'].sum()/scores.shape[0]
```
